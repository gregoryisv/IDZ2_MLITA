<p style="font-size: 17px">9. Исходный код программы.</p>

```python
def read():
    global ch
    ch+=1

def CBE_not_crl():
    if expr[ch] == '[':
        read()
        CBE()
        if expr[ch]==']':
            read()
            CBE_not_sqr()
        else:
            raise ValueError("Expected ] at "+str(ch+1))
    elif expr[ch] == '(':
        read()
        CBE()
        if expr[ch]==')':
            read()
            CBE_not_rnd()
        else:
            raise ValueError("Expected ) at "+str(ch+1))

def CBE_not_rnd():
    if expr[ch] == '[':
        read()
        CBE()
        if expr[ch]==']':
            read()
            CBE_not_sqr()
        else:
            raise ValueError("Expected ] at "+str(ch+1))
    elif expr[ch] == '{':
        read()
        CBE()
        if expr[ch]=='}':
            read()
            CBE_not_crl()
        else:
            raise ValueError("Expected } at "+str(ch+1))

def CBE_not_sqr():
    if expr[ch] == '(':
        read()
        CBE()
        if expr[ch]==')':
            read()
            CBE_not_rnd()
        else:
            raise ValueError("Expected ) at "+str(ch+1))
    elif expr[ch] == '{':
        read()
        CBE()
        if expr[ch]=='}':
            read()
            CBE_not_crl()
        else:
            raise ValueError("Expected } at "+str(ch+1))

def CBE():
    if expr[ch] == '[':
        read()
        CBE()
        if expr[ch]==']':
            read()
            CBE_not_sqr()
        else:
            raise ValueError("Expected ] at "+str(ch+1))
    elif expr[ch] == '(':
        read()
        CBE()
        if expr[ch]==')':
            read()
            CBE_not_rnd()
        else:
            raise ValueError("Expected ) at "+str(ch+1))
    elif expr[ch] == '{':
        read()
        CBE()
        if expr[ch]=='}':
            read()
            CBE_not_crl()
        else:
            raise ValueError("Expected } at "+str(ch+1))
                    

ch=0
expr=input("Enter bracets sequence: ")+"#"

try:
    CBE()
    if expr[ch]=="#":
        print("Every thing is OK")
    else:
        print("Error in outer pair")
except ValueError as ve:
    print(ve)
```
